```python
from scipy.stats import f_oneway
import seaborn as sns
import matplotlib.pyplot as plt


etest_p = dataset['etest_p']
mba_p = dataset['mba_p']

# Perform ANOVA test
statistic, p_value = f_oneway(etest_p, mba_p)

# Set significance level
alpha = 0.05

# Make decision
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in means.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in means.")
```

```python
# 8) Test the similarity between the degree_t(sci&Tech) and specialisation(Mkt&HR)with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)
```

```python
from scipy.stats import ttest_ind



# Select data for Science & Technology degree
science_tech_salary = dataset[dataset['degree_t'] == 'Sci&Tech']['salary'].dropna()

# Select data for Mkt&HR specialization
mkt_hr_salary = dataset[dataset['specialisation'] == 'Mkt&HR']['salary'].dropna()

# Perform independent two-sample t-test
statistic, p_value = ttest_ind(science_tech_salary, mkt_hr_salary)

# Set significance level
alpha = 0.05

# Make decision
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in means.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in means.")
```

```python
#11) Test the similarity between the degree_t(sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)
```

```python
import scipy.stats as stats


data_sci_tech = dataset[dataset['degree_t'] == 'Sci&Tech']

# Calculate the differences
differences = data_sci_tech['etest_p'] - data_sci_tech['mba_p']

# Perform paired sample t-test
t_stat, p_value = stats.ttest_rel(data_sci_tech['etest_p'], data_sci_tech['mba_p'])

# Compare p-value to significance level
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant difference.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference.")
```
